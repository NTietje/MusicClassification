```python
''' 
############### CNN Dropout ################

'''
# define 
folder = 'mirror_cnn_params'
template_path = folder + '/drop/0_1/cnn_drop_0_1_0'

# folders of histories
subfolders = ['drop/0_1','drop/0_2_24','drop/0_3_24','4_300_001_r1'] #'4_300_00025_r1','4_300_0001_r1'

# get best and avgs
folder_hist_list = get_histories_of_subfolders(subfolders, folder)
best, avgs = get_best_and_avgs(folder_hist_list, template_path, epoch_limit=80) #, epoch_limit=80

# graph settings  
labels = ['0_1','0_2_24','0_3_24','none']
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink']

############# avg ###############
plot_graph(
    'val_loss', avgs, labels, 'upper left','Average of 30 Trainings', colors, window=[0.9, 1.4])
plot_graph(
    'val_accuracy', avgs, labels, 'upper left','Average of 30 Trainings', colors, window=[0.65, 0.8])
plot_graph(
    'val_accuracy', avgs, labels, 'upper left','Average of 30 Trainings', colors, window=[0.5, 1.01],  metric2='accuracy')
plot_graph(
    'accuracy', avgs, labels, 'upper left','Average of 30 Trainings', colors, window=[0.2, 1])
bar_plot(
    avgs, labels,'Minimum of 30 Trainings', 'min','val_loss', colors, window=[0, 1.3])
bar_plot(
    avgs, labels,'Test_loss', 'eval','eval_test_loss', colors, window=[0, 1.3])
bar_plot_loss_min(
    avgs, labels,'Val_Accuracy at Minimum Val_Loss','val_accuracy', colors, window=[0, 0.8])
evaluate_bar_plot(avgs, labels, 'accuracy', window=[0, 1], figsize=(7, 1.5))
bar_plot(
    avgs, labels,'Test_acc', 'eval','eval_test_acc', colors, window=[0, 0.8])
bar_plot(
    avgs, labels,'Val_acc', 'eval','eval_val_acc', colors)
bar_plot(
    avgs, labels,'Average of 30 Trainings', 'max','traintime', colors)
```
