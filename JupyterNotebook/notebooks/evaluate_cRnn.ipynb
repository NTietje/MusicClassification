Erstellung der CNN-Darstellungen zu den Trainingsiterationen.
Bei jedem Codeblock verweist der Titel auf den untersuchten Bereich.

```python
from ipynb.fs.full.history_nb import *
from ipynb.fs.full.plot import *
```

Die ersten Bezeichnungen, die in der Ausgabe erscheinen wie "m_cRnn_100_lr00025_23.json", sind die besten Trainingsdurchl√§ufe einer Variante.

```python
''' 
############### CRNN dropout ################

'''
# define 
folder = 'mirror_cRnn_params'
template_path = folder + '/drop/1_1_0/cRnn_drop_1_0_0'

# folders of histories
subfolders = ['drop/0_2_0_24_b','drop/0_3_0_24_b','100_lr00025']

# get best and avgs
folder_hist_list = get_histories_of_subfolders(subfolders, folder)
best, avgs = get_best_and_avgs(folder_hist_list, template_path)

# graph settings  
labels = ['0;2;0-24','0;3;0-24','none']
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:cyan']

############# avg ###############
plot_graph(
    'val_loss', avgs, labels, 'upper left','Average of 30 Trainings', colors, window=[0.9, 1.4])
plot_graph(
    'val_accuracy', avgs, labels, 'upper left','Average of 30 Trainings', colors, [0.5, 0.75])
plot_graph(
    'val_accuracy', avgs, labels, 'upper left','Average of 30 Trainings', colors, [0.5, 1], 'accuracy')
plot_graph(
    'accuracy', avgs, labels, 'upper left','Average', colors, [0.4, 1])
bar_plot(
    avgs, labels,'Minimum of 30 Trainings', 'min','val_loss', colors, [0, 1.3])
bar_plot(
    avgs, labels,'Test_loss', 'eval','eval_test_loss', colors, [0, 1.3])
bar_plot_loss_min(
    avgs, labels,'Val_Accuracy at Minimum Val_Loss','val_accuracy', colors, [0, 1])
evaluate_bar_plot(avgs, labels, 'accuracy', [0, 1], figsize=(7, 1.5))
bar_plot(
    avgs, labels,'Test_acc', 'eval','eval_test_acc', colors, [0.5, 0.8])
```
