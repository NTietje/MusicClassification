```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
```

```python
%matplotlib inline

def elu(z, alpha=1):
    return np.where(z < 0, alpha * (np.exp(z) - 1), z)

def relu(z):
    return np.maximum(0, z)

z = np.linspace(-10, 5)
plt.figure(figsize=(3, 2))
#plt.grid(True)
plt.plot(z, elu(z), linewidth=5, color='tab:orange', label='ELU')
plt.plot(z, relu(z), linewidth=2, color='tab:blue', label='ReLU')
plt.title("ELU and ReLU")
plt.axis([-10, 5, -1, 2])
legend = plt.legend(loc='upper left', prop={'size':12})
plt.savefig("elu_plot")
plt.show()

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)

def sigmoid(z):
    return 1/ (1+np.exp(-z))

x = np.linspace(-10, 10)
plt.figure(figsize=(3, 2))
plt.plot(x, softmax(x), color='tab:blue', linewidth=2)
#plt.grid(True)
plt.axis('tight')
plt.title('Softmax Function')
plt.show()

x = np.linspace(-10, 10)
plt.figure(figsize=(3, 2))
#plt.grid(True)
plt.plot(x, sigmoid(x), color='tab:blue', linewidth=2)
plt.axis('tight')
plt.title('Sigmoid Function')
plt.show()
```

```python
# Plot Mel-Spec
def draw_mel_spec(spec, sr, title ,width=8 , height=3.5):
    fig, ax = plt.subplots()
    fig.set_size_inches(width, height)
    img = librosa.display.specshow(spec, sr=24000, x_axis='time', cmap='magma', y_axis='mel', ax=ax)
    fig.colorbar(img, ax=ax, format='%+2.0f dB')
    
# Darw Waveform
def waveform(path):
        y, sr = librosa.load(path, sr=None, mono=True)
        fig = plt.figure(figsize=(9, 4))
        ax1 = fig.add_subplot(211)
        ax1.set_ylabel('Amplitude')
        librosa.display.waveplot(y, sr=sr)
        fig.subplots_adjust(hspace=.5)
```
